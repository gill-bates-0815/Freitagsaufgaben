# Step1
### Lese die  'dataset.csv' Datei  und speichere den Dataframe in der Variable 'data'. Gib anschließend die ersten 4 Zeilen des Dataframes aus


```python
import pandas as pd
import numpy as np 
data = pd.read_csv("Dataset.csv")
data.head(4)
```

# Step2
## Wie viel NaN hat jeder Spalte!?
### Gib anschließend die 10 größten Werte aus

```python
data.isnull().sum().nlargest(10)
```

# Step 3
### Lösche die Spalten wo die Null Werte 50%>

```python
new_data = data[data.columns[data.isna().sum()/data.shape[0] < 0.5]] # 50% = 0.5
new_data.head(5)
```

# Step 4
### Prozentual gesehen, wie viel NaN Werte gibt es im ganzen DataFrame?

```python
##prozentwert = data.isnull().sum().sum() * 100 / data.size
## prozentwert

prozentwert_bei_50 = new_data.isnull().sum().sum() * 100 / len(new_data)
prozentwert_bei_50
```

# Step6 (recherche Aufgabe)
## Fülle alle NaN Werte in der Spalte "Existing Use" mit dem nächsten darüber liegenden NICHT NaN Wert!
## Im zweiten Schritt, füllt die übrigen NaN Werte mit 0.
### Die beiden folgenden Tabellen sind Beispiele zur Veranschaulichung. Die erste zeigt die Ausgangstabelle und die zweite das Ergebnis nach Anwendung des zweiten Schrittes.

Spalte 1 | Spalte 2 
-------- | -------- 
NaN   | 1  
 3  | NaN 
NaN | 7  
NaN   | NaN 

##### Endergebnis
Spalte 1 | Spalte 2 
-------- | -------- 
0   | 1  
 3  | 1 
3 | 7  
3   | 7 

```python
import pandas
pandas.options.mode.chained_assignment = None
new_data['Existing Use'] = new_data['Existing Use'].fillna(method='ffill') # Änderung des Default-Wertes mit ffill
new_data
```

# Step 7
## Ersetze die NaN Werte in der Spalte "Number of Existing Stories"  mit einer 1

```python
new_data_with_0 = new_data.fillna(0)
new_data_with_0
```

```python
mit_1_gefuellt = new_data["Number of Existing Stories"].fillna(1)
mit_1_gefuellt
```

# Step 8
## Ersetze die NaN Werte in der Spalte "Existing Units"  mit der Maximale Wert Diese Spalte

```python
maximalwert = new_data["Existing Units"].fillna(new_data["Existing Units"].max())
maximalwert
```

# Step 9
## Ersetze die NaN Werte in der Spalte "Plansets"  mit eine Polynomial Interpolation  zweite Ordnung

```python
plansets_graf = new_data.Plansets.interpolate(method= "polynomial", order= 2)
plansets_graf
```
